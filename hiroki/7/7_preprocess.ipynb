```python
# raw data
data_raw = {}

for i in range(8):
  data_raw[f"bus_group{i+1}"] = pd.read_excel(
      "rust_data_clean.xlsx", sheet_name=i
      )
  
data_raw["bus_group3"].head()
```

```python
data_dict = {}

for i in range(8):
  df = data_raw[f"bus_group{i+1}"]
  df = df.replace(0,np.nan)
  #df = df.reset_index()
  df = df.rename(
    index={
        "index": "bus_name",
        0: "month_purchased",
        1: "year_purchased",
        2: "month_1st_rep",
        3: "year_1st_rep",
        4: "odometer_1st",
        5: "month_2nd_rep",
        6: "year_2nd_rep",
        7: "odometer_2nd",
        8: "month_data_begins",
        9: "year_data_begins",
    }
)
  data_dict[f"bus_group{i+1}"] = df

data_dict["bus_group3"]
```

```python
# table2aに必要なデータだけ抽出
# 全てのシートを一つにまとめる

data_table2a = pd.DataFrame()
for i in range(8):
  df = data_dict[f"bus_group{i+1}"][0:10].T
  df.insert(0,"group",str(i+1))
  data_table2a = pd.concat(
      [data_table2a, df],axis = 0
  )

data_table2a
```

```python
# 条件(少なくとも一回エンジンが交換された)データを抽出．
data_table2a = data_table2a.query("year_1st_rep == year_1st_rep")

# Table2aに使う変数を作成．
data_table2a["mileage_2nd"] = data_table2a["odometer_2nd"] - data_table2a["odometer_1st"]
data_table2a["period_2nd"] = (data_table2a["year_2nd_rep"]- data_table2a["year_1st_rep"])*12 + (data_table2a["month_2nd_rep"] - data_table2a["month_1st_rep"]) + 1
data_table2a["period_1st"] = (data_table2a["year_1st_rep"]- data_table2a["year_purchased"])*12 + (data_table2a["month_1st_rep"] - data_table2a["month_purchased"]) + 1

# 必要なデータだけ
data_table2a = data_table2a[["group","odometer_1st","mileage_2nd","period_1st","period_2nd"]]
data_table2a
```

```python
# Table2aに便利な形式に変更
data_table2a_1 = data_table2a[["group","odometer_1st","period_1st"]].set_axis(['group', 'mileage', 'period'], axis=1)
data_table2a_2 = data_table2a[["group","mileage_2nd","period_2nd"]].set_axis(['group', 'mileage', 'period'], axis=1)
data_table2a = pd.concat([data_table2a_1,data_table2a_2],axis = 0)

# 欠損値の削除
data_table2a = data_table2a.dropna()
data_table2a.sort_values("group")
```

```python
# FullSample列を追加
data_table2a["FullSample"] = "Full Sample"
data_table2a_full_sample = data_table2a[["FullSample","mileage","period"]]
data_table2a_full_sample = data_table2a_full_sample.set_axis(['group', 'mileage', 'period'], axis=1)
data_table2a = pd.concat([
    data_table2a, data_table2a_full_sample
])
```

```python
data_table2a.groupby(["group"]).agg(
    {"mileage":["max", "min", "mean", "std","count"]
     ,"period":["max", "min", "mean", "std","count"]}
     ).round(1)
```

## Table2b

```python
data_dict["bus_group3"]
```

```python
data_timeseries_dict = {}
data_year_1st_rep = {}

from dateutil.relativedelta import relativedelta

for i in range(8):
  df = data_dict[f"bus_group{i+1}"].T

  df["year_data_begins"] = df["year_data_begins"] + 1900
  df.insert(10
                      ,"date_data_begins"
                      , pd.to_datetime(
                        df["year_data_begins"].astype(int).astype(str)
                        + "-" 
                        + df["month_data_begins"].astype(int).astype(str)
                        ,format="%Y-%m",)
                        )
  df.insert(11
                      ,"Elapsed_Time",
                      (85-df["year_purchased"])*12 + (5-df["month_purchased"]) + 1
                      )
  # データの条件に使用
  data_year_1st_rep[f"bus_group{i+1}"] = df[["year_1st_rep"]]
  
  df = df.iloc[:,10:]
  data_timeseries_dict[f"bus_group{i+1}"] = df

data_timeseries_dict["bus_group3"].head()
```

```python
# table2b用のデータセットを作成
data_timeseries_renamed_dict = {}
from dateutil.relativedelta import relativedelta

for i in range(8):
  df = data_timeseries_dict[f"bus_group{i+1}"]

  initial_date = df["date_data_begins"].min()
  df = df.drop(columns = {"date_data_begins"})
  
  month_list = [(initial_date + pd.DateOffset(months=j-1)).strftime('%Y-%m') for j in range(len(df.columns))]
  month_list[0] = "Elapsed_Time"

  df.columns = month_list
  df.insert(0, "group", str(i+1))

  data_timeseries_renamed_dict[f"bus_group{i+1}"] = df

data_timeseries_renamed_dict["bus_group3"].head()
```

```python
table2b_dict = {}

#7,8シートは("year_1st_rep == 0")を満たすデータがないので，処理がバグる．rangeの範囲を8から6に変更．
for i in range(6):
  df1 = data_year_1st_rep[f"bus_group{i+1}"]
  df2 = data_timeseries_renamed_dict[f"bus_group{i+1}"]
  table2b_df = pd.concat([df1,df2],axis = 1)
  
  # Table2bの条件に合うデータのみを抽出
  table2b_df = table2b_df.query("year_1st_rep != year_1st_rep")
  table2b_dict[f"bus_group{i+1}"] = table2b_df

table2b_dict["bus_group3"]
```

```python
# Table 2b
# 1985年5月のデータだけを抽出し，全データをマージ．
table2b_df = pd.DataFrame()

for i in range(6):
  df = table2b_dict[f"bus_group{i+1}"]
  table2b_df = pd.concat(
      [table2b_df,df[["group","1985-05","Elapsed_Time"]]]
      ,axis = 0
      )

# FullSampleの追加
table2b_df["FullSample"] = "Full Sample"
table2b_df_full_sample = table2b_df[["FullSample","1985-05","Elapsed_Time"]]
table2b_df_full_sample = table2b_df_full_sample.set_axis(["group","1985-05","Elapsed_Time"], axis=1)
table2b_df = pd.concat([
    table2b_df, table2b_df_full_sample
])
table2b_df = table2b_df.drop(columns = "FullSample")
table2b_df
```

```python
table2b_df.groupby("group").agg(["max", "min", "mean", "std","count"]).round(1)
```

## 遷移行列の推定に用いる，時系列データの作成

```python
rust_timeseries_dict = {}

for i in range(8):
  df = data_timeseries_renamed_dict[f"bus_group{i+1}"]
  df = df.drop(columns=["group","Elapsed_Time"])
  df = df.T
  #df = df.reset_index()
  rust_timeseries_dict[f"bus_group{i+1}"] = df
```
