```python
data = pd.read_csv("data_2.csv")
data
```

5. ロジットモデルの推定

```python
# fixed effect estimation 
# OLS
def create_dot(list_columns):
    for i in list_columns:
        data[f"{i}_dot"] = data[i] - data.groupby(['NameID'])[i].transform('mean')

list_columns = ["price","hppw","FuelEfficiency","size","logit_share"]
create_dot(list_columns)
```

```python
# FE estimation
y = data["logit_share_dot"]
X = data[["price_dot","hppw_dot","FuelEfficiency_dot","size_dot"]]

model_fe = sm.OLS(y,X)
result_fe = model_fe.fit(cov_type = "HC3")

result_fe.summary().tables[1]
```

```python
# Replication:Pooled OLS regression (fixed effect estimationではない！)
y = data["logit_share"]
X = data[["price","hppw","FuelEfficiency","size"]]
X = sm.add_constant(X)

model_ols = sm.OLS(y,X)
result_ols = model_ols.fit(cov_type = "HC3")

result_ols.summary().tables[1]
```

### IV estimation
https://bashtage.github.io/linearmodels/iv/examples/advanced-examples.html

```python
from linearmodels import IV2SLS, IVGMM, IVGMMCUE, IVLIML
```

```python
list_IVs = ['iv_BLP_own_hppw', 'iv_BLP_own_FuelEfficiency',
       'iv_BLP_own_size', 'iv_BLP_other_hppw', 'iv_BLP_other_FuelEfficiency',
       'iv_BLP_other_size', 'iv_GH_own_hppw', 'iv_GH_own_FuelEfficiency',
       'iv_GH_own_size', 'iv_GH_other_hppw', 'iv_GH_other_FuelEfficiency',
       'iv_GH_other_size']
create_dot(list_IVs)
```

```python
IV_BLP = ['iv_BLP_own_hppw_dot', 'iv_BLP_own_FuelEfficiency_dot',
       'iv_BLP_own_size_dot', 'iv_BLP_other_hppw_dot',
       'iv_BLP_other_FuelEfficiency_dot', 'iv_BLP_other_size_dot']
IV_GH = ['iv_GH_own_hppw_dot', 'iv_GH_own_FuelEfficiency_dot',
       'iv_GH_own_size_dot', 'iv_GH_other_hppw_dot',
       'iv_GH_other_FuelEfficiency_dot', 'iv_GH_other_size_dot']
controls = ['hppw_dot', 'FuelEfficiency_dot', 'size_dot']
```

```python
# Fixed effect estimation
# OLS estimation
ols = IV2SLS(data.logit_share_dot, data[["price_dot"] + controls], None, None)
res_ols = ols.fit(cov_type = "robust")

# GMM estimation with BLP IV
iv_gmm_BLP = IVGMM(
    data.logit_share_dot,
    data[controls],
    data.price_dot,
    data[IV_BLP]
)
res_gmm_BLP = iv_gmm_BLP.fit(cov_type = "clustered",
                             clusters = data.NameID)

# GMM estimation with GH IV
iv_gmm_GH = IVGMM(
    data.logit_share_dot,
    data[controls],
    data.price_dot,
    data[IV_GH]
)
res_gmm_GH = iv_gmm_GH.fit(cov_type = "clustered",
                             clusters = data.NameID)

```

```python
from collections import OrderedDict
from linearmodels.iv.results import compare

res = OrderedDict()
res["OLS"] = res_ols
res["gmm_BLP"] = res_gmm_BLP
res["gmm_GH"] = res_gmm_GH
compare(res)
```
